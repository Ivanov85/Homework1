```python
df = pd.read_csv('subscribers.csv')
```

```python
df['user_id'] = pd.Series(range(1000, 2000))
```

```python
cols_1 = ['user_id', 'subscribing_channel', 'date_subscribed', 'is_retained']
df_site = df[cols_1]
df_site.shape
df_site.head()
```

```python
cols_2 = ['user_id', 'date_subscribed', 'date_canceled']
df_date = df[cols_2]
df_date.shape
df_date.head()
```

```python
df_2 = pd.merge(df_date, df_site, on = 'user_id', how = 'left')
df_2.shape
df_2.head()
```
